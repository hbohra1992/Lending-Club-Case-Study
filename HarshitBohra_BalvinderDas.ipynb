```python
# Now Deleting the some more columns which will not contribute much to the analysis 
# pub_rec_bankruptcies ,last_pymnt_d, last_credit_pull_d,collection_recovery_fee,recoveries,collection_recovery_fee
# total_rec_late_fee, total_rec_prncp, out_prncp_inv,out_prncp,pub_rec, mths_since_last_delinq,inq_last_6mths, earliest_cr_line
# delinq_2yrs, zip_code, installment, revol_util

data=data.drop(["pub_rec_bankruptcies","last_pymnt_d","last_credit_pull_d","collection_recovery_fee",
           "recoveries","collection_recovery_fee","total_rec_late_fee",
           "out_prncp_inv","pub_rec","mths_since_last_delinq","inq_last_6mths","earliest_cr_line","delinq_2yrs",
            "zip_code","total_rec_prncp","revol_util","out_prncp"], axis = 1)
```

```python
data.home_ownership.value_counts()
```

```python
# dropping rows for column "home ownership" having None value
data=data[~(data.home_ownership=="NONE")]
```

```python
data.shape
```

```python
# check the values for loan status column
data.loan_status.value_counts()
```

```python
#Removing all the rows having loan status as current
data=data[~(data.loan_status=="Current")]
```

## Analysis using Different type of Visualization

```python
#Loan Status
sns.histplot(data=data.loan_status)
plt.title("Loan Status")
plt.show()
```

###### Count of Fully Paid loans are more than the Charges Off Loans and we have observed that there are around 5000 people who have defaulted the loan

```python
# Checking the Loan amount using dist plot

sns.distplot(x=data.loan_amnt)
plt.title("Loan Amount")
plt.show()
```

###### Most of the People have applied for the loan amount less than or equal to 25000

```python
#Analysis of loan approval ratio
sns.distplot(x=(data.funded_amnt_inv*100)/data.loan_amnt)
plt.title("Approved Loan Amount Ratio")
plt.show()
```

###### approximately 95% loan is approved 

```python
# Checking for loan tenture
data.term=data.term.apply(lambda x:x.replace('months',''))
#sns.histplot(data=data.term)
sns.countplot(x="term",data=data[data.loan_status=="Charged Off"])
plt.title("Loan Term for Charged Off Loan")
plt.show()
```

###### Loan is given for the term of 36 and 60 months and also loan provided for the term of 36 months is more charged off

```python
# lets check the column grade for charged off loans
sns.countplot(x="grade",data=data[data.loan_status=="Charged Off"])
plt.title("Grade of Charged Off Loans")
plt.show()
```

###### From the above count plot it is clear that charged off happens for the people belowng to grade B and C

```python
# Let's check for which subgrade people charged off more
sns.countplot(x="sub_grade",data=data[data.loan_status=="Charged Off"])
plt.title("Sub Grade of Charged Off Loans")
plt.xticks(rotation=90)
plt.show()
```

###### From the above count plot it is clear that charged off happens for the people belongs to sub grade B5

```python
# Check the purpose columns to identify the purpose of loan for which loan status is charged off
sns.countplot(x="purpose",data=data[data.loan_status=="Charged Off"])
plt.title("Purpose of loan whose status is charged off ")
plt.xticks(rotation=90)
plt.show()
```
